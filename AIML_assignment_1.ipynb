```python

arr = np.array([0, np.pi/2, np.pi])

sin_result = np.sin(arr)
print("Sine:", sin_result)

log_result = np.log(arr[arr > 0])
print("Log:", log_result)

log2_result = np.log2(arr[arr > 0])
print("Log2:", log2_result)

exp_result = np.exp(arr)
print("Exp:", exp_result)
```

```python
import pandas as pd
import numpy as np

people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

df = pd.DataFrame({
    'Name': people,
    'Age': age,
    'Height': height,
    'Weight': weight
})
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.scatter(df['Age'], df['Height'], color='blue')
plt.title('Scatter Plot: Age vs Height')
plt.xlabel('Age')
plt.ylabel('Height')

plt.subplot(2, 2, 2)
plt.bar(df['Name'], df['Weight'], color='green')
plt.title('Bar Chart: Weight by Person')
plt.xlabel('Name')
plt.ylabel('Weight')

plt.subplot(2, 2, 3)
plt.hist(df['Height'], bins=5, color='orange', edgecolor='black')
plt.title('Histogram: Height Distribution')
plt.xlabel('Height')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
